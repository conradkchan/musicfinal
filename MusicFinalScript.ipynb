```python
notes = ['a', 'a#', 'b','c', 'c#', 'd', 'd#', 'e', 'f', 'f#', 'g', 'g#']

```

```python
probsDiff = [0.25, 0.11, 0.1, 0.09, 0.08, 0.08, 0.07, 0.07, 0.05, 0.04, 0.03, 0.03]

```

```python
keysDiff = ['c', 'g', 'd#', 'f', 'd', 'a', 'e', 'c#', 'a#', 'g#', 'b', 'f#']
```

```python
def find_note(initial_note, step):
    initial_index = notes.index(initial_note)
    if initial_index == 0 and step == -1:
        return 'g#'
    elif initial_index == 11 and step == 1:
        return 'a'
    else:
        new_index = initial_index + step
        return notes[new_index]
```

```python
def create_song(starting_note, numNotes, aa, ad, da, dd):
    song = []
    song.append(starting_note)
    first_step = np.random.choice([1, -1], 1, p=[0.5, 0.5])[0]
    current_note = find_note(starting_note, first_step)
    song.append(current_note)
    prev_step = first_step
    for i in range(numNotes - 1):
        if prev_step == 1:
            new_step = np.random.choice([1, -1], 1, p=[aa, ad])[0]
        else:
            new_step = np.random.choice([1, -1], 1, p=[da, dd])[0]
        current_note = find_note(current_note, new_step)
        song.append(current_note)
        prev_step = new_step
    return song
```

```python
def convert_song_to_pysynth(song, speed, pr):
    pySong = []
    for note in song:
        pySong.append([note, np.random.choice(speed, 1, p=pr)[0]])
    return pySong
```

```python
starting_note = np.random.choice(keysDiff, 1, replace=True, p=probsDiff)

#Following the average percentagfe of ascending and descending

base_pattern = create_song(starting_note[0], 50, 0.51, 0.49, 0.3, 0.7)

averagePerctSong = convert_song_to_pysynth(base_pattern, [4,8,16], [1/3, 1/3, 1/3])

higherAPercentSong = convert_song_to_pysynth(create_song(starting_note[0], 50, 0.70, 0.3, 0.5, 0.5), [4,8,16], [1/3, 1/3, 1/3])

higherDPercentSong = convert_song_to_pysynth(create_song(starting_note[0], 50, 0.30, 0.7, 0.2, 0.8), [4,8,16], [1/3, 1/3, 1/3])

slower_song = convert_song_to_pysynth(base_pattern, [1,2,4], [1/3, 1/3, 1/3])

faster_song = convert_song_to_pysynth(base_pattern, [8,16,32], [1/3, 1/3, 1/3])

pysynth_s.make_wav(averagePerctSong, fn="averagePerctSong.wav")
pysynth_s.make_wav(higherAPercentSong, fn="higherAPercentSong.wav")
pysynth_s.make_wav(higherDPercentSong, fn="higherDPercentSong.wav")
pysynth_s.make_wav(slower_song, fn="slowerSong.wav")
pysynth_s.make_wav(faster_song, fn="fasterSong.wav")
```
